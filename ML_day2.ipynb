#Importing required libraries

```python
import numpy as np
import pandas as pd
```

#below command fetches datasets from scikit site


```python
from sklearn import datasets 
from sklearn.model_selection import train_test_split
```

#Loading Dataset into variable IRIS


```python
IRIS = datasets.load_iris()
```

# Content Stored in Variable IRIS is displayed

```python
IRIS
```

# We want to view data separately

```python
x=IRIS.data
y=IRIS.target
```

#Output of x and y

```python
print("Input Data",x)
```

##Output of y i.e Target variable

```python
print("Target Data",y)
```

#creating Data Frame

```python
df = pd.DataFrame(data=np.c_[IRIS['data'],IRIS['target']],columns=IRIS['feature_names']+['Species'])
df
```

#To Know dataset mean,std,min,max

```python
df.describe()
```

# Splitting Training and Testing dataset

```python
from sklearn.model_selection import train_test_split
```

```python
# Here test data is of 30% by test_size=0.3 which is in probability and set random state as any number to just fix no of dataset in both
df_train,df_test = train_test_split(df, test_size=0.3, random_state=25)
```

```python
# Here you can see data is randomly separated for training and testing
# Print Train Data
df_train
```

```python
# Print Test Data
df_test
```
